```python

ddd = "2021/3/13 12:00:0"
d = ephem.Date(ddd)
sun.compute(d)
moon.compute(d)
h0 = moon.hlong*180./np.pi

mlist = ['Jan','Feb','Mar','Apr','May','Jun']

for dd in range(240):
    #print(d.tuple())
    sun.compute(d)
    moon.compute(d)
    #print(dd, sun.hlong, moon.hlong, moon.phase)
    d = ephem.Date(d + 6*ephem.hour)
    

    shd = sun.hlong*180./np.pi
    mhd = moon.hlong*180./np.pi
    ddd = d.datetime()
    
    dtup = d.tuple()
    month = dtup[1]
    day = dtup[2]
    hr = dtup[3]
    
    dt = mlist[month-1]
    
    hour = str(hr)
    if hr < 10:
        hour = " " + hour
    s =  "  " +  hour+":00  " + dt + " " + str(day)
    

    diff = mhd - shd  + 180
    diff = (diff + 180) % 360 - 180
    
    longdiff = mhd - h0
    longdiff = (longdiff + 180) % 360 - 180
    
    phase = moon.phase
    print(s, "   %6.1f  %6.1f  %6.1f   %6.2f "% (  (shd+180)%360, longdiff%360, diff, phase))


```
