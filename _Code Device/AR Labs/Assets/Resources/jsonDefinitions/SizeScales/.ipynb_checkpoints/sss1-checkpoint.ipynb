```python
names = {}

fn1 = 'asu-names-new.tsv'
f1= open(fn1,"r")
lines1 = f1.readlines()
for i in range(49, len(lines1)-1):
    l = lines1[i]
    vv = l.strip().split("\t")
    names[int(vv[0])] = vv[1]

f1.close()
```

```python
# filter only the brightest stars - add the names when possible
magMax = 2.78
ct = 0
distMax = 1500.0


starList = []

for i in range(64,len(lines)-1):
#for i in range(64,70):
    vv = lines[i].strip().split("\t")
    #print(vv)
    raD = float(vv[0])
    decD = float(vv[1])
    hr = int(vv[2])
    greek = vv[4]
    cons = vv[5]
    dist = float(vv[10])
    mag = float(vv[11])
    spec = vv[12]
    if mag < magMax and dist < distMax:
        
        try:
            name = names[hr]
        except:
            name = ""
            
       # if dist > dmax:
       #     dmax = dist
        starList.append(
            {"name":name,
            "ra":raD,
            "dec":decD,
            "dist":dist,
            "mag":mag,
            "hr":hr,
            "greek":greek,
            "const":cons,
            "spec":spec}
        )
        #print(name, vv[2], dist, mag, spec)
        ct = ct + 1
        
#print('ctotal = ',ct)
```

```python
simScale = 1.0/ 500.0

x1 = []
y1 = []
z1 = []
for s in starList:
    theta = s['ra'] * np.pi/ 180.0
    phi = s['dec']* np.pi/ 180.0
    dist = s['dist']
    
    r = dist * simScale
    xx = r * np.cos(theta) * np.sin(phi)
    yy = r * np.sin(theta) * np.sin(phi)
    zz = r * np.cos(phi)
    x1.append(xx)
    y1.append(yy)
    z1.append(zz)
```

```python
importantNames = ['Algol', 'Alcor','Altair','Antaries','Arcturus','Alioth','Betelgeuse','Rigel','Capella','Canopus','Deneb','Fomalhaut','Polaris','Sirius','Thuban','Spica','Vega', 'Pollux']

importantNames = ['Algol', 'Altair','Antares','Arcturus','Betelgeuse','Rigel','Capella','Deneb','Polaris','Sirius','Spica','Vega', 'Pollux']

invertNames = {}
for k in names.keys():
    v = names[k]
    invertNames[v] = k

lessNames = {}
for nn in importantNames:
    lessNames[invertNames[nn]] = nn
```

```python
olist = []
otype= "Prefabs/tinySphere"
oscale = 0.01
parent = "[_DYNAMIC]"
vmin = -1



for s in starList:
    
    if s['name'] != "":
        sname = s['name']
    else:
        sname = s['greek'] + " " + s['const']

    theta = s['ra'] * np.pi/ 180.0
    phi = s['dec']* np.pi/ 180.0
    dist = s['dist']
    color = [255,255,255,255]
    r = dist * simScale
    xx = r * np.cos(theta) * np.sin(phi)
    yy = r * np.sin(theta) * np.sin(phi)
    zz = r * np.cos(phi)
    
    
    distPC = s['dist'] /3.26
    vmag = s['mag']
    vMag = vmag - 2.5 * np.log10(distPC/10)

    
#    dv = 10**((vMag - vmin)/2.5)  # the actual ratio for magnitudes
    dv = 10**((vMag - vmin)/4.)  # the formula we are using to keep things easier to see
    if dv < 1:
        dv = 1
    brightness = int(255 / dv)
    
    
    color = [255, 255, 255, brightness]
    newStar =  {'type':otype,
                'position': {'x': xx, 'y': yy, 'z': zz},
                'scale': {'x': oscale, 'y': oscale, 'z': oscale},
                'parentName': parent,
                'color': color,
                'name': sname}
    olist.append(newStar)
#    print(json.dumps(newStar,indent=4))


```

```python
olist2 = []

otype = "Prefabs/textPrefab"
oscale = 0.1
fSize = 1.0
textColor = [255,255,255,255]

for k in lessNames.keys():
    nn = lessNames[k]
       # print(nn)
    sid = [s for s in starList if s['name'] == nn]
    
    if len(sid) == 1:
        #print(sid)
        parent = sid[0]['name']
        sname = 'Label'+parent
        xx = 0.0
        yy = 0.1
        zz = 0.0
        
        ra = sid[0]['ra']
        dec = sid[0]['dec']
        
        tdata = {
            "textField": parent,
            "color": textColor,
            "fontSize": fSize,
            "wrapText": False
            }

        newLabel =  {'type':otype,
                    'tmp':json.dumps(tdata),
                    'position': {'x': xx, 'y': yy, 'z': zz},
                    'scale': {'x': oscale, 'y': oscale, 'z': oscale},
                    'eulerAngles': {'x': dec, 'y': ra, 'z': 0.0},
                    'parentName': parent,
                    'name': sname}
        olist2.append(newLabel)
        #print(json.dumps(newLabel, indent=4))
```
