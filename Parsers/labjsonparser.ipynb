# Lab Json Parsing

## Break Down JSON into parts
break into modules, then each modules clips, then all the object changes in the clips

```julia
labjson = JSON.parsefile("Full_Lab_Transmission.json")
modulesjson = [JSON.parse(activityjson) for activityjson in labjson["ActivityModules"]]
clips = [[clip for clip in mod["clips"]] for mod in modulesjson]
obj_changes = [[[modification for modification in clip["objectChanges"]] for clip in mod["clips"]] for mod in modulesjson]
```

```julia
# function that takes a list and prints it in a grid, with a specified number of columns
# Read left to right
function printgrid(list, cols)
    # Get the number of rows
    rows = Int(ceil(length(list)/cols))
    # Construct the list of each column
    colslist = [list[i:cols:end] for i in 1:cols]
    # find length of longest string in each column and add 2
    colwidths = [maximum(length.(string.(col))) + 2 for col in colslist]
    # Print the grid with Padding
    for i in 1:rows
        for j in 1:cols
            index = (i-1)*cols + j
            if index <= length(list)
                item = string(list[index])
                print(item * " "^(colwidths[j] - length(item)))
            end
        end
        println()
    end
end
```

```julia
unique_names = []
for i in 1:length(obj_changes)
    for j in 1:length(obj_changes[i])
        for k in 1:length(obj_changes[i][j])
            if obj_changes[i][j][k]["name"] ∉ unique_names
                push!(unique_names, obj_changes[i][j][k]["name"])
            end
        end
    end
end

printgrid(unique_names, 5)
```
